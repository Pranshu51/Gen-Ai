```python
data1 = pd.DataFrame(
    data=np.random.randint(1, 100, size=(100, 4)),
    columns=list('abcd')
)  # This parenthesis closes pd.DataFrame()
```

```python
data2 = pd.DataFrame(data=np.random.randint(1,100,size=(100,3)), columns=list('xyz'))
```

```python
data1
```

```python
data2
```

```python
combined_data = pd.concat([data1, data2], axis=0)  # Stack vertically (row-wise)
```

```python
combined_data  #column ka bhi addition and row kka bhi additon means 100+100=200 and 4+3=7
```

```python
combined_data = pd.concat([data1, data2], axis=1) 
combined_data
#only column ka ddtion hoga no rows addiotn
```
